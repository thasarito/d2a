```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd

# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('all.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df = df[df['startelo32'].notnull()]
```

```python
df = df[df['bday'].notnull()]
```

```python
df['bday'] = pd.to_datetime(df['bday'], format="%Y-%m-%d")
```

```python
df['start'] = pd.to_datetime(df['start'], format="%Y-%m-%d")
df['end'] = pd.to_datetime(df['end'], format="%Y-%m-%d")
```

```python
df['ageattour'] = df['start'] - df['bday']
```

```python
df['elo32diff'] = df['endelo32'] - df['startelo32']
```

```python
df[['elo32diff', 'ageattour']]
```

```python
df.dtypes
```

```python
df['ageattour'] = (df['ageattour'] / np.timedelta64(1, 'D')).astype(int)
```

```python
y = df['startelo32']
x = df['ageattour']
plt.scatter(x, y)
```

```python
avgarteezy = (df[df['player'] == 'Arteezy'].startelo32 + df[df['player'] == 'Arteezy'].endelo32).mean() / 2
avgarteezy
```

```python
stdarteezy = ((df[df['player'] == 'Arteezy'].startelo32 + df[df['player'] == 'Arteezy'].endelo32) / 2).std()
```

```python
df[df['player'] == 'Arteezy'].startelo32 = (df[df['player'] == 'Arteezy'].startelo32 - avgarteezy )/ avgarteezy
```

```python
arteezyseries = df[df['player'] == 'Arteezy'].startelo32
```

```python
df[df['player'] == 'Arteezy']['normal'] = (arteezyseries - avgarteezy)/ stdarteezy
```

```python
df
```

```python
df.loc[df['player'] == 'Arteezy', 'normal'] = (arteezyseries - avgarteezy)/ stdarteezy
```

```python
df2 = df[df['player'] == 'Arteezy']
```

```python
x = df2['ageattour']
y = df2['normal']
plt.scatter(x, y)
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 3))(np.unique(x)))
```

```python
cumsums = df[['player', 'start', 'elo32diff']].groupby(['player', 'start']).sum().groupby(level=0).cumsum()
```

```python
df2 = df.set_index(['player', 'start'])
```

```python
df2['cumelo'] = cumsums
```

```python
df2 = df2.reset_index()
```

```python
df2
```

```python
x = df2['ageattour']
y = df2['cumelo']
plt.scatter(x, y, alpha = 0.5)
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 2))(np.unique(x)), 'r')
```
